Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import numpy as np
import pandas as pd
```

```python
df = np.random.randint(1,10,(200,3))
```

```python
df = pd.DataFrame(df,columns=['A','B','C'])
```

```python
df
```

```python
import pandas as pd

def re_index(df):
    new_col = pd.Index(range(1,len(df)*2,2))
    df['new_col'] = new_col
    df.set_index('new_col',inplace=True)
    return df
```

```python
re_index(df)
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
df = pd.date_range(start='2022-01-01',end='2024-01-01')
```

```python
df
```

```python
df = pd.to_datetime(df)
```

```python
df = pd.DataFrame({'Date':df})
```

```python
df['Date']
```

```python
df[(df['Date']>='2023-01-01') & (df['Date']<='2023-01-31')]
```

```python
def calling(df,start,end):
    df1 = df[(df['Date']>=start) & (df['Date']<=end)]
    return df1
```

```python
calling(df,'2023-01-01','2023-01-31')
```
