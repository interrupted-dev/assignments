Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math
mean = 0
std = 1
x= 0.5
exponent = -((x - mean)**2)/(2*std**2)
coeff = 1/(std*math.sqrt(2*math.pi))
pdf = coeff*math.exp(exponent)
pdf
```

```python
def n_pdf(mean,std,x):
    exponent = -((x - mean)**2)/(2 * std ** 2)
    coefficient = 1/(std*math.sqrt(2*math.pi))
    pdf = coefficient * math.exp(exponent)
    return pdf
```

```python
from scipy.stats import norm
norm.pdf(0.5, loc=0, scale=1)
```

```python
n_pdf(0,1,0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean = 1000
std = 5
x = np.linspace(mean-4*std,mean+4*std,1000)
y = norm(mean,std).pdf(x)
```

```python
plt.plot(x,y,c="blue",label=f"μ={mean},σ={std}" )
plt.axvline(mean,ls='--',color="red")
plt.xlabel("x_values")
plt.ylabel("y values(pdf)")
plt.title("pdf of normal distribution")
plt.grid(True)
plt.legend()
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

#parameter
sample_size=1000
probability_of_success=0.4
no_of_trials=1
random_sample = binom.rvs(n=no_of_trials,p=probability_of_success,size=sample_size)
plt.hist(random_sample,edgecolor='black')
plt.title('binomial distribution')
plt.xlabel("no of success")
plt.ylabel("frequency")

          
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
from scipy.stats import poisson
def cal_cdf(mean,point):
    return poisson.cdf(point, mu=mean)
```

```python
cal_cdf(3,5)
```

```python
#and for pmf
import math
def poisson_pmf(x,mean_rate):
    return (math.exp(-mean_rate)*(mean_rate**x))/math.factorial(x)
```

```python
poisson_pmf(5,3)
```
